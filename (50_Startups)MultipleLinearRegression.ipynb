```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

df=pd.read_csv("50_Startups.csv")
df
```

```python
x=df.iloc[:,:-1].values  #parameters in x   ,result in y
y=df.iloc[:,4].values
```

```python
print(x,y)   #parameters are independent so taking as x and y
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
Label=LabelEncoder()
x[:,3]=Label.fit_transform(x[:,3])
```

```python
print(x)
```

```python
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([('encoder',OneHotEncoder(),[3])],remainder='passthrough')          #onehotencoder assigns values(0,1,2)
x=np.array(ct.fit_transform(x))
print(x)
```

```python
print(pd.DataFrame(x).head())
```

```python
from sklearn.model_selection import train_test_split
train_test_split(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=14)
```

```python
#LinearRegression
print(x.shape)
print(y.shape)
print(x_test.shape)
print(y_test.shape)
print(x_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
l=LinearRegression()
l=l.fit(x_train,y_train)
l
```

```python
y_pred=l.predict(x_test)
y_pred
```

```python
y_test
```

```python
result=pd.concat([pd.DataFrame(y_pred,columns=['Prediction']),
                 pd.DataFrame(y_test,columns=['Test'])],axis=1)
result['Difference']=result['Prediction']-result['Test']
result
```

```python
r2_score(y_test,y_pred)
```

```python
y_final_pred=l.predict(x)
y_final_pred
```

```python
#final prediction
y_fpred=pd.DataFrame(y_final_pred,columns=['Prediction'])
result=pd.concat([df,y_fpred],axis=1)
result['Difference']=result['Profit']-result['Prediction']
result
```

```python
r2_score(df['Profit'],y_final_pred)
```
