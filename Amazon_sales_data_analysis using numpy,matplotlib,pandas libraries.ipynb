calculate the number of lower and upper case alphabets in a given string

```python
def calc_string(x):
  u_count=0
  l_count=0
  for char in x:
    if char.isupper():
      u_count+=1
    elif char.islower():
      l_count+=1
  return u_count,l_count
print(calc_string("She Sells Seashells by the Seashore"))
```

```python
#solve 2x+y=8  ,3x+4y=18 using matrix

m1=np.array([[2,1],[3,4]])
co_m2=np.array([8,18])

solution=np.linalg.solve(m1,co_m2)
print(solution)
```
