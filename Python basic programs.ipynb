```python
def recursum(n):
  if n<=1:
    return n
  return n+recursum(n-1)
n=10
print(recursum(n))
```

```python
#Fibonacci series
#1
def rfib(n):
  if n<=1:
    return(n)
  else:
     return(rfib(n-1)+rfib(n-2))
n=4
for i in range(n):
  print(rfib(i))
```

```python
#2
def recursive_fibonacci(n):
    if n <= 1:
        return n
    else:
        return(recursive_fibonacci(n-1) + recursive_fibonacci(n-2))
n_terms = 5
if n_terms <= 0:
    print("Invalid input ! Please input a positive value")
else:
    print("Fibonacci series:")
for i in range(n_terms):
    print(recursive_fibonacci(i))
```

```python
def recur_fact(n):
  if n==1:
    return(n)
  else:
    return n*recur_fact(n-1)
n=5
print(recur_fact(n))
```

```python
def gen_num(n):
  v=0
  while v<n:
    yield v
    v+=1
for v in gen_num(100):
  print(v)
```
