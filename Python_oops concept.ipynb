```python
#1(2)
class shopping:
  def __init__(self):
    self.items=[]
    self.cost=[]
  def shop_items(self,item):
    s=item
    self.items.append(s)
  def cost_item(self,b):
    c=b
    self.cost.append(c)

  def total_bill(self):
    total=0
    for i in self.cost:
      total=total+i
    return(total)
```

```python
x=shopping()
```
