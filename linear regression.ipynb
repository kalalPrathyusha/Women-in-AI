Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("/content/citizen_pci.csv")
df
```

```python
plt.scatter(df.year,df.pci)
plt.xlabel("Year")
plt.ylabel("pci")
plt.show()
```

```python
plt.plot(df.year,df.pci)
plt.xlabel("Year")
plt.ylabel("pci")
plt.show()
```

```python
import sklearn.linear_model as linear_model             #sklearn=scikit-learn  contains all the  models and libraries
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df.pci)
```

```python
reg
```

```python
reg.predict([[2030]])
```

```python
#1
plt.scatter(df.year,df.pci)
plt.plot(df.year,reg.predict(df['year']),color='red')
plt.xlabel("Year")
plt.ylabel("pci")
plt.show()
```

```python
#1(1)
plt.scatter(df.year,df.pci)
plt.plot(df.year,reg.predict(df[['year']]),color='red')
plt.xlabel("Year")
plt.ylabel("pci")
plt.show()
```
